```python
categories = ['Laptop', 'Smart TV', 'Smartphone', 'Tablet']
values = [307, 305, 321, 310]
colors = ['#4c72b0', '#55a868', '#c44e52', '#8172b3']
```

```python
max_value_index = values.index(max(values))
explode = [0.1 if i == max_value_index else 0 for i in range(len(categories))]
plt.figure(figsize=(8, 6))  # Set the figure size for a beautiful layout
plt.pie(values, labels=categories, colors=colors, explode=explode, autopct='%1.1f%%', startangle=140,)
plt.axis('equal')
plt.title('Devices Utilization')
```
